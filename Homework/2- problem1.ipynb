```python
#List of characters
disney_character = ["simba","ariel","pumba","flounder","nala","ursula","scar","flotsam","timon"]
#Produce loops
for x in disney_character:
    if "u" in x:
        print(x, "U are so uniquely U!")
    elif "i" in x:
        print(x, "I bet you're Impressively Intelligent!")
    elif "o" in x:
        print(x, "O my! How original!")
    else:
        print(x, "Ehh, a's and e's are so ordinary.")
```
