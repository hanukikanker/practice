```python
# My favorite romance movies
# title, release year, runtime, tagline, main characters
romantic_movie1 = ("The Princess Bride", 1987, 98, "The story of a man and a woman who lived happily ever after.", ["Buttercup", "Westley", "Fezzik", "Inigo Montoya","Vizzini"])
romantic_movie2 = ("Groundhog Day", 1993, 101, "He's having the day of his life...over and over again.",["Phil Connors"])
romantic_movie3 = ("Amélie", 2001, 122, "One person can change your life forever.",["Amélie Poulain", "Nino Quincampoix", "The Garden Gnome"])
movie_list = [romantic_movie1,romantic_movie2,romantic_movie3]
```

```python
#1. Print out elements within tuples
print('Here are my favourite romance movies:\n ')
for movie in movie_list:
    print(f'{movie[0]} ({movie[1]}): {movie[3]}\n')
```

```python
# Just poracticing different things!
# 2. Trying to make a function to take in how many movies to list out:
def fav_movie():
    # User input
    n = input('How many movies do you want to list out? ')
    # Print out
    if int(n) <= len(movie_list):
        # Messages
        if int(n) == len(movie_list):
            print(f'Here are all of my favourite romance movies:\n ')
        elif int(n) == 1:
            print(f'Here is one of my favourite romance movies:\n ')
        else:
            print(f'Here are {n} of my favourite romance movies:\n ')
        #  Movie Details
        for i in range(int(n)):
            print(f'{movie_list[i][0]} ({movie_list[i][1]}): {movie_list[i][3]}\n')
    # Print out 'error'
    else:
        print(f'There are only {len(movie_list)} movies in your library\n')
        fav_movie()
```
